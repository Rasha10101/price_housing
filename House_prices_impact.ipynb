```python
sns.set_theme(style="darkgrid")
```

```python
colors = ["pink", "lightblue" ,"silver","lavender"]
for col in cat.columns:
    plt.figure(figsize=(6, 6))
    data[col].value_counts().plot.pie( autopct='%1.1f%%',startangle=90,colors=colors,wedgeprops={"linewidth":3 })
    plt.title(f"Pie Chart for {col}")
    plt.ylabel('')
    plt.show()
```

# **نقوم بمقارنة الهدف مع جميع الفئات**

# **Barplot**

## لمقارنة الاسعار مع كل فئة

```python
overall_average_price = data['price'].mean()
for col in cat.columns:
    plt.figure(figsize=(10, 4))
    sns.barplot(data=data, x=col, y='price', palette="twilight_shifted")
    plt.axhline(y=overall_average_price, color='red', linestyle='--', label='Overall Average Price')
    plt.title(f'Comparison of {col} with price')
    plt.ylabel('Average Price')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()
```

# **تشفير الاعمدة الغير رقمية**

```python
from sklearn.preprocessing import LabelEncoder

non_num = df.select_dtypes(exclude=['float64', 'int64']).columns
label_encoder = LabelEncoder()
for col in non_num :
    df[col] = label_encoder.fit_transform(df[col])
print(df.head())
```
